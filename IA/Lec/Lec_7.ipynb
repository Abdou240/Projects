```python
#Importing our dataset from csv file
import pandas as pd
dataset=pd.read_csv("titanic-passengers.csv", sep=";")
# Head of the Dataset
dataset.head(2)
#Titanic.tail()
```

```python
dataset.isnull().sum()
```

```python
#Preprocessing our data
def preprocess_data(new_data):
    new_data['Age'].fillna(new_data['Age'].mean(),inplace=True) # deal with missing values of Age Variable
    new_data.replace({'Sex':{'male': 1,'female':0}},inplace=True) # convert the categories of the Variable Sex
    new_data['Cabin']=new_data.Cabin.fillna('G6') # replace missing Catogieries of the Varibale "Cavin"
    new_data.replace({'Survived':{'Yes': 1,'No':0}},inplace=True) # replace missing Catogieries of the Varibale Survived
    return new_data
data=preprocess_data(dataset)
```

```python
data
```

```python
data.isnull().sum()
```

```python
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#features extraction
x=data.drop(["PassengerId","Survived", "Name", "Cabin", "Ticket", "Embarked"], axis=1)
y= data["Survived"]
```

```python
#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=50)
#applying tree algorithm
dtree = tree.DecisionTreeClassifier() # Build decision Model  
dtree.fit(x_train, y_train)   #fitting our model
y_pred=dtree.predict(x_test)   # evaluating our model  /Testing model`s Performance
print("score:{}".format(accuracy_score(y_test, y_pred))) # Print Acurency
```

```python
from sklearn import datasets
#from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier #Importing Random Forest Classifier
from sklearn.model_selection import cross_val_score 
from sklearn import metrics  # Importing metrics to test accuracy

iris = datasets.load_iris()

clf=RandomForestClassifier(n_estimators=100)  #Creating a random forest with 100 decision trees
scores = cross_val_score(clf,iris.data,iris.target,cv = 5)
print("Accuracy on each fold/subset :", scores)
print("The avrage accurency is ", sum(scores)/5)
```
