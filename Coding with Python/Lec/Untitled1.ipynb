```python
# import pandas library
import pandas as pd
  
# dictionary with list object in values
details = {
    'Name' : ['Ankit', 'Aishwarya', 'Shaurya',
              'Shivangi', 'Priya', 'Swapnil'],
    'Age' : [23, 21, 22, 21, 24, 25],
    'University' : ['BHU', 'JNU', 'DU', 'BHU', 
                    'Geu', 'Geu'],
}
  
# creating a Dataframe object 
df = pd.DataFrame(details, columns = ['Name', 'Age',
                                      'University'],
                  index = ['a', 'b', 'c', 'd', 'e', 'f'])
print(df)
# get names of indexes for which
# column Age has value 21
index_names = df[ df['Age'] == 21 ].index
print(index_names)
# drop these row indexes
# from dataFrame
df.drop(index_names, inplace = True)
  
df
```

```python
import pandas as pd
# used for np.nan
import numpy as np
# declared dictionary
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']} 
print(exam_data)
# DataFrame without using labels
Df = pd.DataFrame(exam_data)
print (Df)
# DataFrame using labels
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
Df.index = labels 
# another Methode Df = pd.DataFrame(exam_data,index =labels)
print(Df)
print ('#####################1################################')
# 1- Print the three first rows using head() method
print(Df.head(3))
print ('#####################2################################')
#Delete rows with Nan values
# index of Df['score'].isnull() == True
index_score = Df[ Df['score'].isnull() == True ].index
print(index_score)
# 2 -drop these row indexes from dataFrame
Df.drop(index_score, inplace = True)
print(Df)
print ('#####################3################################')
# 3- Extract the 'name' and 'score' columns from the DataFrame.
print(Df[['name','score']])
print ('#####################4################################')
#4- Write a Pandas program to append a new row 'k' to data frame with these values(name : "Suresh", score: 15.5, attempts: 1, qualify: "yes")
#exam_data = {'name': ['Anastasia'
exam_data2 = {'name' : "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
print(exam_data2)
Df2 = pd.DataFrame(exam_data2,index =['k'])
New_Data= pd.concat([Df,Df2])
print(New_Data)
print ('#####################5################################')
# 5- Write a Pandas program to delete the 'attempts' column from the DataFrame.
New_Data1 =New_Data.drop('attempts',axis=1)
print(New_Data1)
```
